# LIMPEZA E TRANSFORMAÃ‡ÃƒO


# MODELAGEM

```python
modelo = LinearRegression()
```

```python
X = df_historico[['clarity', 'carat','cut','color','clarity','depth','table','x','y','z']] # Ã© um dataframe => VARIAVEIS EXPLICATIVAS
y = df_historico['price'] #Ã© uma serie
```

```python
modelo.fit(X,y)
```

```python
coeficiente = modelo.coef_[0]
coeficiente
```

```python
intercepto = modelo.intercept_
intercepto 
```

# PREVISÃƒO

```python
dado_para_prever = df_estimar[['clarity', 'carat','cut','color','clarity','depth','table','x','y','z']]
dado_para_prever
```

```python
y_pred = modelo.predict(dado_para_prever)
y_pred
```

```python
df_estimar['price_predicted'] = y_pred
```

```python
df_estimar.to_csv('todas_numericas.csv', index=False)
```
