# MODELAGEM

```python
modelo = LinearRegression()
```

```python
X = df_historico[['carat']] # é um dataframe => VARIAVEIS EXPLICATIVAS
y = df_historico['price'] #é uma serie
```

```python
modelo.fit(X,y)
```

```python
coeficiente = modelo.coef_[0]
coeficiente
```

```python
intercepto = modelo.intercept_
intercepto 
```

# PREVISÃO

```python
dado_para_prever = df_estimar[['carat']]
dado_para_prever
```

```python
y_pred = modelo.predict(df_estimar[['carat']])
y_pred
```

```python
df_estimar['price_predicted'] = y_pred
```
